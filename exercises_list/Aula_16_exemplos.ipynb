## Comparando com o primeiro número
```python
try:
    lista = [int(input()) for _ in range(10)]
    ref = lista[0]
    maiores = sum(x > ref for x in lista)
except:
    print('Erro ao processar a lista.')
```


## Verificar Primo
```python
def verificar_primo(n):
    try:
        n = int(n)
        for i in range(2, int(n**0.5)+1):
            if n % i == 0:
                return False
        return True
    except ValueError:
        print('Número inválido')
```
