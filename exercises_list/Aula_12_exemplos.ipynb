# Aula 12 - Exemplos com Strings


## Exemplo - Índices e Fatiamento
```python
texto = 'Sorria! Hoje é segunda!'
print(texto[0:6])
print(texto[::-1])  # Inverter
```


## Exemplo - Substituição com Replace
```python
frase = 'Eu gosto de Python'
frase_nova = frase.replace('Python', 'programar')
print(frase_nova)
```


## Exemplo - Verificando Palíndromo
```python
texto = 'A sacada da casa'.replace(' ', '').lower()
if texto == texto[::-1]:
    print('É palíndromo')
else:
    print('Não é palíndromo')
```


## Exemplo - Contagem de Caracteres
```python
frase = 'Olá mundo'
print('Espaços:', frase.count(' '))
print('A:', frase.lower().count('a'))
```
