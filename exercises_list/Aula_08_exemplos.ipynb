# Aula 08 - Exemplos de Estruturas de Repetição


## Exemplo 1 - Contador Simples
```python
for i in range(5):
    print('Execução número', i)
```


## Exemplo 2 - While com Condição
```python
numero = 0
while numero < 5:
    print('Número é:', numero)
    numero += 1
```


## Exemplo 3 - Acumulador
```python
soma = 0
for i in range(1, 6):
    soma += i
print('Soma total:', soma)
```


## Exemplo 4 - Usando math.sqrt()
```python
import math
delta = 16
raiz = math.sqrt(delta)
print('Raiz quadrada:', raiz)
```


## Exemplo 5 - Laços Aninhados (Tabuada)
```python
for i in range(1, 11):
    for j in range(1, 11):
        print(f'{i} x {j} = {i*j}')
    print('------')
```
