# Aula 14 - Manipulação de Arquivos (Parte 1)


## Exercício 1
Cadastre o nome e a raça de 5 cachorros. Grave no arquivo `.txt`, uma linha para cada cachorro.


## Exercício 2
Cadastre N pessoas (nome, sobrenome, idade) e salve em arquivo. A 1ª linha indica quantos foram cadastrados.


## Exercício 3
Leia os dados dos cachorros do arquivo `.txt` e exiba na tela.
