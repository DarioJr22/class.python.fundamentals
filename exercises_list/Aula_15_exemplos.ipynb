# Exemplos - Aula 15 - Profundidade e Votação


## Medições maiores que a anterior
```python
with open('medicoes.txt') as f:
    medidas = [int(l) for l in f]
count = sum(medidas[i] > medidas[i-1] for i in range(1, len(medidas)))
print(count)
```


## Votação
```python
with open('votos.txt', 'w') as f:
    for _ in range(10):
        voto = input('1-Sushi, 2-Churrasco: ')
        f.write(voto + '\n')
```
