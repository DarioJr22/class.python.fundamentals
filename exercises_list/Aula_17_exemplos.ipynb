# Exemplos - Aula 17 - Try/Except/Else


## Exceção com else
```python
try:
    x = int(input())
except ValueError:
    print('Erro')
else:
    print('Número digitado foi:', x)
```


## Verificar Primo com else
```python
def primo(n):
    try:
        n = int(n)
    except:
        print('Erro')
    else:
        print('É primo' if all(n % i != 0 for i in range(2, n)) else 'Não é primo')
```
