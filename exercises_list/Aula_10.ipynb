# Aula 10 - Matrizes


## Exercício 1
Crie um programa com um menu CRUD para uma lista de nomes:
- Create, Read, Update e Delete
- Verifique se o índice é válido antes de Update e Delete


## Exercício 2
Crie uma matriz 3x3. Solicite os valores e, depois:
- Calcule a soma de todos os números ímpares
- Calcule a soma dos valores da primeira coluna
- Encontre o menor valor da terceira linha
