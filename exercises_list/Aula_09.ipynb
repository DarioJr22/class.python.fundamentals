# Aula 09 - Listas (Vetores)


## Exercício 1
Solicite 5 palavras ao usuário, armazene em uma lista e exiba em ordem alfabética.


## Exercício 2
Peça a quantidade de produtos vendidos no dia, leia o peso de cada um, e calcule:
- Peso médio
- Maior e menor peso
- Total arrecadado (1kg = R$4,35)


## Exercício 3
Leia 10 números inteiros, armazene-os em uma lista. Depois, separe os números pares e ímpares em listas diferentes e exiba as três listas.


## Exercício 4
Solicite as notas de 5 alunos, calcule a média da turma e mostre as notas acima da média.
