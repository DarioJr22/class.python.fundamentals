# Aula 16 - Tratamento de Erros


## Exercício 1
Tenha um vetor de 10 números. Conte quantos são maiores, menores e iguais ao primeiro. Trate possíveis erros.


## Exercício 2
Crie a função `verificar_primo(numero)` com tratamento de exceções.


## Exercício 3
Abra um arquivo e, se ele não existir, avise o usuário com try/except.
