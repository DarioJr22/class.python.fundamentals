# Exemplos - Aula 14 - Arquivos


## Escrever cachorros
```python
nomes = ['Paçoca', 'Rex']
racas = ['Labrador', 'Poodle']
with open('cachorros.txt', 'w') as f:
    for nome, raca in zip(nomes, racas):
        f.write(f'Nome: {nome} Raça: {raca}\n')
```


## Ler arquivo
```python
with open('cachorros.txt', 'r') as f:
    for linha in f:
        print(linha.strip())
```
