# Aula 11 - Matrizes (Continuação)


## Exercício 1
Receba uma matriz 3x3 e calcule:
- Produto dos elementos da diagonal principal
- Média de todos os elementos
- Maior valor da segunda coluna
- Quais valores são menores ou iguais à média


## Exercício 2
Crie um programa que leia a quantidade de garrafas vendidas em cada região do Brasil, em cada trimestre do ano (matriz 4x5), e exiba o total vendido no país.

💡 *Dica:* Use uma matriz onde cada linha representa um trimestre e cada coluna representa uma região.
