# Exemplos - Aula 13 - Funções e Dicionários


## Função com parâmetros
```python
def intervalo(lista, lim_inf, lim_sup):
    return [x for x in lista if lim_inf <= x <= lim_sup]
```


## Dicionário com quadrados
```python
n = 5
d = {i: i**2 for i in range(n+1)}
print(d)
```


## Cardápio com dicionário
```python
cardapio = {'Café': 4.5, 'Pão': 2.0}
cardapio['Suco'] = 5.0
if 'Pão' in cardapio:
    cardapio['Pão'] = 2.5
```
