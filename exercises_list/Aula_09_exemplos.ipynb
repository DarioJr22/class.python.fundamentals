# Aula 09 - Exemplos com Listas


## Exemplo 1 - Criando e Acessando uma Lista
```python
frutas = ['maçã', 'banana', 'laranja']
print(frutas[0])
```


## Exemplo 2 - Adicionando e Ordenando
```python
numeros = []
numeros.append(5)
numeros.append(3)
numeros.sort()
print(numeros)
```


## Exemplo 3 - Encontrando Máximo, Mínimo e Soma
```python
valores = [10, 20, 30]
print('Maior:', max(valores))
print('Menor:', min(valores))
print('Soma:', sum(valores))
```


## Exemplo 4 - Separando Pares e Ímpares
```python
pares = []
ímpares = []
numeros = [1, 2, 3, 4, 5]
for n in numeros:
    if n % 2 == 0:
        pares.append(n)
    else:
        ímpares.append(n)
print('Pares:', pares)
print('Ímpares:', ímpares)
```
