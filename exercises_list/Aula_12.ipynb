# Aula 12 - Strings


## Exercício 1
Peça uma frase, uma palavra existente nela e uma nova palavra. Substitua e exiba a frase modificada em maiúsculas.

💡 *Dica:* Verifique se a palavra existe antes com 'in'.


## Exercício 2
Leia um nome e imprima-o na vertical em formato de escada, usando letras maiúsculas.


## Exercício 3
Verifique se uma frase é um palíndromo, desconsiderando espaços.

💡 *Exemplos:* 'A sacada da casa', 'arara'.


## Exercício 4
Conte em uma frase:
- Quantos espaços em branco há
- Quantas vogais (a, e, i, o, u) existem
