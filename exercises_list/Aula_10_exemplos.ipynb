## Exemplo 1 - Menu CRUD Simples
```python
nomes = []
while True:
    print('1-Adicionar, 2-Listar, 3-Atualizar, 4-Remover, 5-Sair')
    opcao = int(input('Escolha: '))
    if opcao == 1:
        nomes.append(input('Nome: '))
    elif opcao == 2:
        for i, nome in enumerate(nomes):
            print(i, nome)
    elif opcao == 3:
        i = int(input('Índice: '))
        if 0 <= i < len(nomes):
            nomes[i] = input('Novo nome: ')
    elif opcao == 4:
        i = int(input('Índice: '))
        if 0 <= i < len(nomes):
            nomes.pop(i)
    elif opcao == 5:
        break
```


## Exemplo 2 - Criando uma Matriz
```python
matriz = []
for i in range(3):
    linha = []
    for j in range(3):
        linha.append(int(input(f'Valor para [{i}][{j}]: ')))
    matriz.append(linha)
print(matriz)
```
