#### 1. Variáveis e Tipos de Dados
### Python usa tipagem dinâmica:

```python
# Declaração de variáveis
nome = "Ana Paula"
altura = 1.70
idade = 30
genero = 'F'

# Verificação de tipos
print(type(nome))    
print(type(altura))  
print(type(idade))    
print(type(genero))   
```

```python
# Entrada de dados com conversão
nome = input("Digite seu nome: ")
cidade = input("Digite sua cidade: ")
idade = int(input("Digite sua idade: "))
salario = float(input("Digite seu salário: "))

print(f"\nTipo de nome: {type(nome)}")
print(f"Tipo de idade: {type(idade)}")
```

```python
# Operadores Aritméticos
a = 10
b = 3
print(f"Soma: {a + b}")         
print(f"Subtração: {a - b}")    
print(f"Multiplicação: {a * b}") 
print(f"Divisão: {a / b:.2f}")   
print(f"Divisão Inteira: {a // b}") 
print(f"Resto: {a % b}")        
print(f"Potência: {a ** b}")     

# Operadores Relacionais
print(f"Maior que: {a > b}")     
print(f"Igualdade: {a == 10}")   

# Operadores Lógicos
x = True
y = False
print(f"E lógico: {x and y}")   
print(f"OU lógico: {x or y}")   
```

#### Exercício Prático: Cálculo de Viagem

```python
distancia = float(input("Distância total (km): "))
velocidade = float(input("Velocidade média (km/h): "))
consumo = float(input("Consumo médio (km/l): "))
preco_combustivel = float(input("Preço do litro: R$"))

# Cálculos
tempo_viagem = distancia / velocidade
litros_necessarios = distancia / consumo
custo_total = litros_necessarios * preco_combustivel

# Resultados
print("\n--- RESULTADOS ---")
print(f"Tempo de viagem: {tempo_viagem:.1f} horas")
print(f"Combustível necessário: {litros_necessarios:.1f} litros")
print(f"Custo total: R${custo_total:.2f}")
```

### Exercício Prático: Salário de Vendedor

```python
# Entrada de dados
carros_vendidos = int(input("Carros vendidos: "))
valor_vendas = float(input("Valor total das vendas: R$"))
salario_fixo = float(input("Salário fixo: R$"))
comissao_por_carro = float(input("Comissão por carro: R$"))

# Cálculo salário final
salario_final = salario_fixo + (comissao_por_carro * carros_vendidos) + (valor_vendas * 0.05)

# Resultado
print(f"\nSalário final: R${salario_final:.2f}")
```
